```python
# Dropping rows where ESRB rating is missing since that's our target variable
data = data.dropna(subset=['esrb_rating'])

# Mapping ESRB ratings to integers
esrb_mapping = {'Everyone': 0, 'Everyone 10+': 1, 'Teen': 2, 'Mature': 3, 'Adults Only': 4}
data['esrb_rating'] = data['esrb_rating'].map(esrb_mapping)
```

```python
# Creating dummy variables for the 'genres' column
data2 = get_dummies(data, ['genres'], dtype='int')
```

```python
x=data2.drop('esrb_rating', axis =1)# feature
y=data2['esrb_rating']# label
print(x.shape[1])
print(y.shape)
```
